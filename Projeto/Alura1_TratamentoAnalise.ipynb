```python
data = [7,2,4,4,5]
s = pd.Series(data)
s
```

```python
s = pd.Series(data = data, index = index)
s
```

```python
#Mesmo processo porém utilizando dicionários
data = {'Linha' + str(i) : i + 2 for i in range(5) }
data
```

# Extra: Organizando Dataframes

```python
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
```

```python
list('321')
```

```python
dataframe1 = pd.DataFrame(data, list('321'), list('ZYX'))
dataframe1
```

```python
#Método que organiza o dataframe pelo índice de linhas
#Para que o método sort_index() modifique o DataFrame, o argumento inplace deve ser modificado para True
dataframe1.sort_index(inplace = True)
dataframe1
```

```python
#Invertendo a ordem pelo label da linha e coluna
dataframe1.sort_index(inplace = True, axis = 1)
dataframe1
```
